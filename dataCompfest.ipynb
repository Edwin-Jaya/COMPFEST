```python
a=data_subset.drop(data_subset.index[177:])
b=a.drop(a.index[1:162])
newData=b.reset_index()
print(newData)
```

```python
newData.columns
```

```python
dataBaru=newData[['Tanggal','Kasus Positif Aktif',
       'Kasus Positif Aktif.1', 'Kasus Positif Aktif.2']]
dataBaru
```

```python
df2 = dataBaru.set_axis(['Tanggal', 'Total Kasus Aktif', 'PPLN', 'NON-PPLN'], axis=1, inplace=False)
df2=df2.drop(df2.index[0:1])
df2
```

### ChartBar Graphic

```python
ppln_list = df2['PPLN'].to_list()
non_ppln_list = df2['NON-PPLN'].to_list()

timestamp_list = df2['Tanggal'].to_list()
date_list = [dt.strftime("%d-%m-%Y") for dt in timestamp_list]

chart_df = pd.DataFrame({'PPLN': ppln_list, 'NON-PPLN': non_ppln_list}, index=date_list)
chart_df.plot.bar(stacked=True, title="stacked chart bar")
```

```python
chart_df.plot.bar(subplots=True, title="subplots chart bar")
```
